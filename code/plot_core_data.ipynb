```python
from string import ascii_letters

import pandas as pd
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

from cgeniepy.data import foram_names, filter_foramdf
from cgeniepy.plot import scatter_map
from cgeniepy.utils import mean_w_na
```

# MARGO data

## LGM

```python
lgm_data = pd.read_csv("~/lgm_ocn/data/lgm_foram_count_data/LGM_MARGO_grouped.csv")

foram_name = list(foram_names().keys())

fig, axs = plt.subplots(2, 2, subplot_kw={"projection": ccrs.Robinson()}, figsize=(8, 5))

for i, ax in enumerate(axs.flat):
    df = filter_foramdf(lgm_data, foram_group=foram_name[i])
    # replace NA to zero and calculate mean
    mean_proportion = mean_w_na(df.Prop,"zero")
    p = scatter_map(ax=ax, df=df, var="Prop", s=20, vmax=1.0, cmap="viridis")
    ax.set_title(f"({ascii_letters[i]}) {foram_name[i]}: global mean={mean_proportion: .1%}")

# Left, bottom, width, length
cbar_ax = fig.add_axes([0.95, 0.165, 0.025, 0.7])
cbar = fig.colorbar(p, cax=cbar_ax, extend="max")
cbar.set_label('Relative abundance', fontsize=11)
```

# Combination of EPILOG (S. Atlantic) and CLIMAP (the rest)

```python
epilog = pd.read_csv("~/lgm_ocn/data/lgm_foram_count_data/LGM_EPILOG_groupped.csv")
climap = pd.read_csv("~/lgm_ocn/data/lgm_foram_count_data/LGM_CLIMAP_groupped.csv")

foram_name = list(foram_names().keys())

fig, axs = plt.subplots(2, 2, subplot_kw={"projection": ccrs.Robinson()}, figsize=(8, 5))

for i, ax in enumerate(axs.flat):
    df1 = filter_foramdf(epilog, foram_group=foram_name[i])
    df2 = filter_foramdf(climap, foram_group=foram_name[i])
    p = scatter_map(ax=ax, df=df1, var="Relative_Abundance", s=20, vmax=1.0, cmap="viridis")
    p = scatter_map(ax=ax, df=df2, var="Relative_Abundance", s=20, vmax=1.0, cmap="viridis")

# Left, bottom, width, length
cbar_ax = fig.add_axes([0.95, 0.165, 0.025, 0.7])
cbar = fig.colorbar(p, cax=cbar_ax, extend="max")
cbar.set_label('Relative abundance', fontsize=11)
```

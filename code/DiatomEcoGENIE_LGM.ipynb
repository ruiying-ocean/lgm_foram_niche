```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

from cgeniepy.core import GenieModel as gm

fig, axes = plt.subplots(nrows=3, ncols=3,
                         figsize=(9, 6),
                         subplot_kw=dict(projection=ccrs.EckertIV()),
                         constrained_layout=True)

diatom_lst = [f"eco2D_Plankton_C_{i:03}" for i in range(1,4)]

# MODELS
mod_pi = gm("../model/Hol_diat.Aaron23")
mod_lgm = gm("../model/LGM_diat.Aaron23")

diatom_lst = [f"eco2D_Plankton_C_{i:03}" for i in range(1,4)]
size_lst = [2, 20, 200]

for i, ax in enumerate(axes.flat):
    if i % 3 == 0:
        p = mod_lgm.select_var(diatom_lst[i//3]).plot_map(ax=ax, cbar=False, vmin=0, vmax=0.5, cmap="viridis")
        ax.set_title(f"LGM, diatom {size_lst[i//3]} $\mu$m")
    if i % 3 == 1:
        p = mod_pi.select_var(diatom_lst[i//3]).plot_map(ax=ax, cbar=False, vmin=0, vmax=0.5, cmap="viridis")
        ax.set_title(f"Holocene, diatom {size_lst[i//3]} $\mu$m")
    if i % 3 == 2:
        p2 = mod_lgm.diff(mod_pi, diatom_lst[i//3]).plot_map(ax=ax, cbar=False, vmin=-0.5, vmax=0.5)
        ax.set_title(f"LGM - Hol, diatom {size_lst[i//3]} $\mu$m")

fig.colorbar(p, ax=axes[:, :-1], shrink=0.4, orientation="horizontal", pad=0.05)
fig.colorbar(p2, ax=axes[:, -1], shrink=0.8, orientation="horizontal", pad=0.05)
fig.savefig("../output/LGM_diatom_diff.jpg", dpi=300)
```

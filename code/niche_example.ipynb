```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(10000)

# Generate random data for species X
num_points_X = 500
temperature_X = np.random.normal(3, 3, num_points_X)
chl_concentration_X = np.random.normal(10, 3, num_points_X)
species_X = np.repeat('X', num_points_X)

# Generate random data for species Y
num_points_Y = 400
temperature_Y = np.random.normal(15, 3, num_points_Y)
chl_concentration_Y = np.random.normal(5, 2, num_points_Y)
species_Y = np.repeat('Y', num_points_Y)

# Create a DataFrame for the dataset
df = pd.DataFrame({'Temperature': np.concatenate([temperature_X, temperature_Y]),
                   'Chl Concentration': np.concatenate([chl_concentration_X, chl_concentration_Y]),
                   'Species': np.concatenate([species_X, species_Y])})

# Add labels and title
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Chl Concentration (mg/m3)')
plt.title('2D Niche example')

## set font family
plt.rcParams['font.family'] = 'Roboto'

# Create the 2D density plot
sns.kdeplot(data=df, x='Temperature', y='Chl Concentration', hue='Species', shade=True, fill=True)

# Save the plot
plt.savefig('../output/niche.png', dpi=300)

# Show the plot
plt.show()
```
